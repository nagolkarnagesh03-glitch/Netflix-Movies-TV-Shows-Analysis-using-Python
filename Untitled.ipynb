```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("netflix.csv")
```

```python
df.columns
```

```python
# DATA CLAENING PROCESS 
```

```python
df.head()
```

```python
df.shape
```

```python
df.dtypes
```

```python
df.isna().sum() / len(df) * 100
```

```python
df.drop_duplicates
```

```python
df.isna().sum()
```

```python
df["director"] = df["director"].fillna("Unknown")
```

```python
df["cast"] = df["cast"].fillna("Unknown")
```

```python
df["country"] = df["country"].fillna("Unknown")
```

```python
df["rating"] = df["rating"].fillna(df["rating"].mode()[0])
```

```python
df.dropna(subset=["date_added"] ,inplace = True)
```

```python
df.shape
```

```python
df.duplicated().sum()
```

```python
df.columns
```

```python
df['date_added'] = pd.to_datetime(df['date_added'])

df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month_name()
df['day_added'] = df['date_added'].dt.day
```

```python
df.to_csv("Netflix.csv" , index = False)
```

```python
# Exploratory Data Analysis (EDA)
```

```python
df.dtypes
```

```python
# How many Movies and TV Shows are available on Netflix?
df["type"].value_counts()

df["type"].value_counts().plot(kind="barh")
plt.title("Movies vs TV shows")
plt.xlabel("Types")
plt.ylabel("Count")
plt.show()
```

```python
# Which content rating is most common?
df["rating"].value_counts()

df["rating"].value_counts().plot(kind="barh")
plt.title("Content Rating Distribution")
plt.xlabel("Ratings")
plt.ylabel("No of Titles")
plt.show()
```

```python
# Which countries produce the most Netflix content?
df["country"].value_counts().head(10)

df["country"].value_counts().head(10).plot(kind="barh")
plt.title("country wise produce content")
plt.xlabel("country")
plt.ylabel("Titles")
plt.show()
```

```python
# Who are the top directors on Netflix?
df["director"].value_counts().head(10)

df["director"].value_counts().head(10).plot(kind="barh")
plt.title("Top directos")
plt.ylabel("Directors Name")
plt.show
```

```python
# Which genres appear most frequently?
df["genres"].value_counts().head(10)

df["genres"].value_counts().head(10).plot(kind="barh")
plt.title("Top Genres")
plt.show()
```

```python
df.columns
```

```python
# Which year had the highest number of titles added to Netflix?
df["date_added"].value_counts().sort_index()

df["date_added"].value_counts().sort_index().plot(kind="barh")
plt.title("Content added by year")
plt.xlabel("Years")
plt.ylabel("Titles Added")
plt.show()
```

```python
# Which release years have the most titles?
df["release_year"].value_counts().sort_index()

df["release_year"].value_counts().sort_index().plot(kind="line")
plt.title("Release years Distributions")
plt.show()
```

```python
df.columns
```

```python
# Which countries produce the most Movies?
df.groupby("country") ["title"].count().sort_values(ascending=False).head().plot(kind="barh")
plt.title("country")
plt.show()
```

```python
# How many titles were released before 2000?
df[df["release_year"] < 2000].shape[0]
```

```python
# What are the 10 newest titles released?
df.sort_values(by="release_year" , ascending=False)[["title" , "release_year"]].head(10)
```

```python
df.columns
```

```python
# How many titles were added to Netflix in each month?
df["month_added"].value_counts().sort_index().plot(kind="bar")
plt.title("monthly titles")
plt.xlabel("month")
plt.ylabel("titles")
plt.show()
```

```python
# Which month has the highest number of content additions?
df["month_added"].value_counts().idxmax()
```
